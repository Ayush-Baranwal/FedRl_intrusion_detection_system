```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid') # darkgrid, white grid, dark, white and ticks
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('font', size=13)          # controls default text sizes
```

```python
df1 = pd.read_csv('rl.csv')
df2 = pd.read_csv('fed4.csv')
df3 = pd.read_csv('fed5.csv')
df4 = pd.read_csv('fed6.csv')
df1 = df1.iloc[1: , :]
df2 = df2.iloc[1: , :]
df3 = df3.iloc[1: , :]
df4 = df4.iloc[1: , :]
print(len(df1))
print(len(df2))
print(len(df3))
print(len(df4))
# print(df1.head(2))
# print(df1.iloc[:, 0])
# print(df1.head(1))
df1 = df1[['avg_attacker_episode_rewards','avg_defender_episode_rewards','hack_probability','cumulative_hack_probability','avg_episode_steps']]
df2 = df2[['avg_attacker_episode_rewards','avg_defender_episode_rewards','hack_probability','cumulative_hack_probability','avg_episode_steps']]
df3 = df3[['avg_attacker_episode_rewards','avg_defender_episode_rewards','hack_probability','cumulative_hack_probability','avg_episode_steps']]
df4 = df4[['avg_attacker_episode_rewards','avg_defender_episode_rewards','hack_probability','cumulative_hack_probability','avg_episode_steps']]
```

```python
df1['index_column'] = list(range(0, 9900, 100))
df2['index_column'] = list(range(0, 9900, 100))
df3['index_column'] = list(range(0, 9900, 100))
df4['index_column'] = list(range(0, 8900, 100))
```

```python
plt.figure(figsize=(10,6), tight_layout=True)
#plotting
plt.plot(df1['index_column'], df1['avg_attacker_episode_rewards'], 'o-', linewidth=2, color=sns.color_palette('bright')[0])
plt.plot(df2['index_column'], df2['avg_attacker_episode_rewards'], 'o-', linewidth=2, color=sns.color_palette('bright')[1])
plt.plot(df3['index_column'], df3['avg_attacker_episode_rewards'], 'o-', linewidth=2, color=sns.color_palette('bright')[2])
plt.plot(df4['index_column'], df4['avg_attacker_episode_rewards'], 'o-', linewidth=2, color=sns.color_palette('bright')[3])
#customization
# plt.xticks(df1['index_column'])
plt.xticks(np.arange(0, 10000+1, step=1000))

plt.xlabel('No of iterations')
plt.ylabel('avg_attacker_episode_rewards')
plt.title('avg_attacker_episode_rewards vs No of iterations')
plt.legend(title='Agent', title_fontsize = 13, labels=['RL agent','Federated agent-1','Federated agent-2','Federated agent-3',])
plt.show()
```
